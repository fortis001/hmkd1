## [과제 ] 고객 구매데이터 생성(id, 성별, 연령, 지역, 상품, 가격, 구매시간) 하고 데이터 프레임으로 변환한 후 아래 사항을 수행하세요.

#### 구매 데이터 생성
- id : 1번 ~ 10000번 일련번호(관측치 10000개)
- gender : 0,1 정수 난수 생성
- age : 10 ~ 90사이 정수 난수를 생성
- region : 1 ~ 10 사이 정수 난수 생성
- pdt_22, pdt_23 : 5개 제품 코드 (11 ~ 55)
- pa_22, pa_23 : 1000 ~ 50000 사이 난수 생성
- q_22, q_23 : 1 ~ 100 사이 난수 생성
- t_22, t_23 : 01 ~ 24 사이 정수 난수 생성('가장 자주 구매한 시간대)
- channel : 구매한 매장 유형(1 ~ 5) 5개 정수 난수 생성
- tr_y : 고객의 구매 이력(누적연수 1 ~ 10)
- 구매액은 price * qty로 산출('19년, '20년 각각 계산)하여 pur_19, pur_20 컬럼으로 생성 
- sales: 구매액 증감을 1, 감소를 0으로 표기
- np.nan으로 각 컬럼별로 10개씩 null값을 삽입하고 각 칼럼의 특성을 고려한 null 값 처리

#### 인사이트 도출
- 통계적 분석 및 시각화하여 탐색한 후 불필요한 컬럼은 삭제하고 필요시 추가 파생변수 생성한 후 분석용 데이터셋을 작성하고 인사이트 3개 이상 도출
- 인사이트 도출하지 못한 경우 데이터를 수정하여 다시 탐색한 후 인사이트 도출 및 기술

#### 데이터 수정시 참조
- 구매액 증감에 영향을 주는 주요 요인과 관련된 인사이트를 만들 수 있는 데이터로 변경하는 점에 중점을 둠
- 구매 감소한 고객의 특성으로 성별, 지역, 선호제픔, 선호 구매시간대, 평균 구매가 level(저가, 중가, 고가)별로 데이터 수정
- 연령, 성별로 상품유형, 구매단가, 구매시간대 등을 복합적으로 그룹화하여 데이터 특성 부여
- 연도별로 상기한 사항들의 변동과 구매 감소간의 상관성을 부여하는 데이터 수정
- 데이터 수정은 불린 인덱싱으로 특정 그룹을 선택한 후 고객 특성에 맞는 값으로 변경

```python
def ctage(age):
    cat = ""
    if age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'
    return cat


genderu = {0 : {"kpop" : 50,"발라드" : 50,"팝송": 20,"클래식": 20,"힙합/락": 10},
           1 : {"kpop": 50,"발라드": 50,"팝송": 20,"클래식": 20,"힙합/락" : 100}}
ageu = {'Child' : {"kpop" : 100,"발라드": 40,"팝송": 20,"클래식" : 10,"힙합/락": 30},
        'Teenager' : {"kpop" : 110,"발라드": 70,"팝송": 40,"클래식" : 20,"힙합/락": 120},
        'Student' : {"kpop" : 120,"발라드": 100,"팝송": 50,"클래식" : 30,"힙합/락": 100}, 
        'Young Adult' : {"kpop" : 130,"발라드": 130,"팝송": 70,"클래식" : 40,"힙합/락": 90},
        'Adult' : {"kpop" : 110,"발라드": 100,"팝송": 70,"클래식" : 70,"힙합/락": 40},
       'Elderly' : {"kpop" : 70,"발라드": 100,"팝송": 50,"클래식" : 150,"힙합/락": 0}}
channelu = {"멜론" : {"kpop" : 11.91,"발라드": 16.92,"팝송": 0.1254,"클래식" : 0.3135,"힙합/락": 10.03},
            "지니" : {"kpop" : 13.62,"발라드": 16.87,"팝송": 0.7788,"클래식" : 0.3245,"힙합/락": 10.38},
            "벅스" : {"kpop" : 11.50,"발라드": 17.89,"팝송": 0.1917,"클래식" : 0.3834,"힙합/락": 10.86},
            "네이버 뮤직" :  {"kpop" : 18.70,"발라드": 17.36,"팝송": 0.2004,"클래식" : 0.6012,"힙합/락": 11.35},
            "Flo" : {"kpop" : 17.60,"발라드": 22.88,"팝송": 0.0044,"클래식" : 0.0044,"힙합/락": 13.20}}

customer_id = np.arange(10000)
gender = np.random.randint(0,2,10000)
age = np.random.randint(10,81,10000)
age_cat = []
i = 0
while i < 10000:
    age_cat.append(ctage(age[i]))
    i += 1
region = np.random.randint(1,6,10000)

pdt = ["kpop","발라드","팝송","클래식","힙합/락"]
pdt_22 = np.random.choice(pdt, 10000)
pdt_23 = np.random.choice(pdt, 10000)
price_22 = np.random.randint(1000,50001,5)
pa_dic_22 = {"kpop" : price_22[0], "발라드" : price_22[1], "팝송" : price_22[2], "클래식" : price_22[3], "힙합/락" : price_22[4]}
pa_dic_23 = {"kpop" : price_22[0]+np.random.randint(-1000,1000)
             , "발라드" : price_22[1]+np.random.randint(-1000,1000)
             , "팝송" : price_22[2]+np.random.randint(-1000,1000)
             , "클래식" : price_22[3]+np.random.randint(-1000,1000)
             , "힙합/락" : price_22[4]+np.random.randint(-1000,1000)}
i = 0
pa_22 = []
while i < 10000:
    pa_22.append(pa_dic_22.get(pdt_22[i]))
    i += 1
i = 0
pa_23 = []
while i < 10000:
    pa_23.append(pa_dic_23.get(pdt_23[i]))
    i += 1
  

char = np.random.randint(1,6,10000)
chad = {1 : "멜론",
          2: "지니",
          3: "벅스",
          4: "네이버 뮤직",
          5: "Flo"}
channel = []
i = 0
while i < 10000:
    channel.append(chad.get(char[i]))
    i += 1


q_22 = []
q_23 = []
i = 0
while i < 10000:
    a = np.round(np.clip(np.random.normal(genderu[gender[i]][pdt_22[i]] + ageu[age_cat[i]][pdt_22[i]] + channelu[channel[i]][pdt_22[i]]
                                 ,3,1),0,np.inf))
    q_22.append(int(a))
    i += 1
i = 0
while i < 10000:
    a = np.round(np.clip(np.random.normal(genderu[gender[i]][pdt_23[i]] + ageu[age_cat[i]][pdt_23[i]] + channelu[channel[i]][pdt_23[i]]
                                 ,3,1),0,np.inf))
    q_23.append(int(a))
    i += 1


t_22 = np.random.randint(1,25,10000)
t_23 = np.random.randint(1,25,10000)

                   
tr_y = np.random.randint(1,11,10000)




columns = ['id','gender','age','region','pdt_22',"pdt_23",'price_22',"price_23",'qty_22',"qty_23",'time_22',"time_23",
          "channel","tr_y","pur_22","pur_23","sales"]
data = {'id' : customer_id,'gender': gender,'age' : age,"age_cat" : age_cat,'region' : region,'pdt_22' : pdt_22,"pdt_23" : pdt_23,
        'price_22' : pa_22,"price_23" : pa_23,'qty_22' : q_22,"qty_23" : q_23, 
        'time_22' : t_22,"time_23" : t_23, "channel" : channel,"tr_y" : tr_y}





```

```python
df = pd.DataFrame(data)
df = df.set_index("id")
```

```python
df["pur_22"] = df["price_22"] * df["qty_22"]
df["pur_23"] = df["price_23"] * df["qty_23"]
```
