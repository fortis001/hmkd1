### 과제1_0502
"https://n.news.naver.com/mnews/article/032/0003161880?sid=101" 에서

연예, 스포츠, 날씨, 프리미엄을 출력하시오.

```python
url = "https://n.news.naver.com/mnews/article/032/0003161880?sid=101"
req = requests.get(url, headers = headers)
soup = BeautifulSoup(req.content, "html.parser")
lines = soup.select("body > div > header > div > div.Ngnb > div > div > div.Ngnb_left > div span.Nicon_family")
for i in lines:
    print(i.text)
```

### 과제 2_0502
- 구글에서 "빅데이터" 라는 검색어로 검색한 후 검색 결과를 가져와서 출력하고 파일로 저장하시오.
- 네이버 홈페이지에서 원하는 내용을 조회해서 그 결과를 출력하고 파일로 저장하시오

```python
with open("google.txt", "w") as f:
    f.write("")
```

```python

driver = webdriver.Chrome("C:/tool/chromedriver.exe")
driver.get("https://www.google.com/")

time.sleep(1)

search_box = driver.find_element(By.NAME, "q")

search_box.send_keys("빅데이터")

search_box.send_keys(Keys.ENTER)

time.sleep(1)

elements = driver.find_elements(By.XPATH, '''//*[@id="rcnt"]''')

for element in elements:
    print(element.text)
    print(element.text, file = open("google.txt", "a", encoding = "utf-8"))
    
time.sleep(1)
driver.close()
```

### 과제 3_0502

URL = "https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=102" 에서

실시간 뉴스 제목을 가져와서 출력하시오. 단, 최신 버전이어야 함.

```python
URL = "https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=102"
driver = webdriver.Chrome("C:/tool/chromedriver.exe")
driver.get(URL)

time.sleep(1)

elements = driver.find_elements(By.CLASS_NAME, 'cluster_text_headline')

for element in elements:
    print(element.text)
    
time.sleep(1)
driver.close()
```

### 과제 4_0502

"https://www.naver.com/" 뉴스 > 사회 카테고리를 클릭하고 뉴스 제목을 출력하시오.

```python
URL = "https://www.naver.com/"
driver = webdriver.Chrome("C:/tool/chromedriver.exe")
driver.get(URL)

time.sleep(1)

news_button = driver.find_element(By.XPATH, '''//*[@id="NM_FAVORITE"]/div[1]/ul[2]/li[2]/a''')

news_button.click()

social_button = driver.find_element(By.XPATH, '''/html/body/section/header/div[2]/div/div/div[1]/div/div/ul/li[4]/a/span''')

social_button.click()

elements = driver.find_elements(By.CLASS_NAME, 'cluster_text_headline')

for element in elements:
    print(element.text)
    
time.sleep(1)
driver.close()
```

### 과제5_0502

한빛출판네트워크 사이트에 자동 로그인하고 마일리지 점수를 가져와서 출력하시오.

```python
URL = "https://www.hanbit.co.kr/"
driver = webdriver.Chrome("C:/tool/chromedriver.exe")
driver.get(URL)

time.sleep(1)

login_button = driver.find_element(By.XPATH, '''//*[@id="wrap_nav"]/ul[2]/li[1]/a''')

login_button.click()

ID_input = driver.find_element(By.XPATH, '''//*[@id="m_id"]''')
PW_input = driver.find_element(By.XPATH, '''//*[@id="m_passwd"]''')

ID_input.send_keys("abcdef")
time.sleep(0.5)
PW_input.send_keys("123456")
PW_input.send_keys(Keys.ENTER)

my_button = driver.find_element(By.XPATH, '''//*[@id="wrap_nav"]/ul[2]/li[3]/a''')
my_button.click()

element = driver.find_element(By.XPATH, '''//*[@id="container"]/div/div[2]/dl[1]/dd/span''')

print(f"나의 마일리지 : {element.text}점")

time.sleep(1)
driver.close()
```
