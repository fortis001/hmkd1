```python
import re
import urllib
import requests
from bs4 import BeautifulSoup
```

### 과제 1_0501.

"https://news.naver.com/" 에서 title 내용을 세 가지 방법으로 출력하시오

```python
url = "https://news.naver.com/"
html = urllib.request.urlopen(url)
bs = BeautifulSoup(html, "lxml")
bs
```

```python
print(bs.find("title").text)
print(bs.find(id = "browserTitleArea").text)
print(bs.select_one("title").text)
```

### 과제2_0501
"http://www.naver.com" 에서 span 태그에 연결된 한글만을 불필요한 공백을 제거한 후 출력하시오.

```python
url = "http://www.naver.com"
res = requests.get(url).text
bs = BeautifulSoup(res, "html.parser")

sps = bs.find_all("span")
" ".join(re.findall("[가-힣]+", str(sps)))
```

### 과제3_0501
"http://www.naver.com" 에서 span 태그에 연결된 문자 + 숫자를 불필요한 공백을 제거한 후 출력하시오.

```python
url = "http://www.naver.com"
res = requests.get(url).text
bs = BeautifulSoup(res, "html.parser")

sps = bs.find_all("span")
" ".join(re.findall("[가-힣\d]+[가-힣]", str(sps)))
```

### 과제4_0501
"http://www.naver.com" 사이트에서 a 태그에 id가 있는 경우에 대하여 연결된 문자 + 숫자를 아래와 같이 출력하시오.
- 해당되는 조건에 맞게 모두 출력
- 한글만 열방향으로 출력
- 불필요한 공백을 제거한 후 한글만 한 줄에 출력

```python
url = "http://www.naver.com"
res = requests.get(url).text
bs = BeautifulSoup(res, "html.parser")

sps = bs.find_all("a", id = True)
print(sps)
result = re.findall("[가-힣\d]+[가-힣]", str(sps))
```

```python
for i in result:
    print(i)
```

```python
print(" ".join(result))
```

### 과제 5
네이버 영화 랭킹 실시간 조회 데이터를 가져와서 Top10영화를 아래와 같이 출력하시오.

1: 던전 앤 드래곤 : 도적들의 명예

2: 존 윅

3: 존 윅 3: 파라벨룸

```python
url = "https://serieson.naver.com/v3/movie/ranking/realtime"

r = requests.get(url)
html = r.text
bs = BeautifulSoup(html, "html.parser")
result = bs.find_all('span',class_="Title_title__s9o0D")

i = 0
while i < 10:
    print(f"{i+1}. : {result[i].text}")
    i += 1
```
