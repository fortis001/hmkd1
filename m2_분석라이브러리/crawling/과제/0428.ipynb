### 과제2_0428

url = "https://news.naver.com" 에서 문서를 가져와서 한글만 출력하시오. (최대한 텍스트 형태로)
- requests 사용
- urllib 사용

```python
url = "https://news.naver.com"
req = requests.get(url)
soup = BeautifulSoup(req.text, "html.parser")
soup = str(soup)
kor = re.findall("[가-힣]+", soup)
result = " ".join(kor)
result
```

```python
response = urllib.request.urlopen(url)
soup = BeautifulSoup(response, "html.parser")
soup = str(soup)
kor = re.findall("[가-힣]+", soup)
result = " ".join(kor)
result
```

```python
for location in soup.find_all('location'):
    province = location.find('province')
    if province and province.text == '강원도영서':
        children = location.find_all(recursive=False)
        for child in children:
            print(re.sub('<[^>]+>', '', str(child)))
            print("-"*50)
            
```

```python
for location in soup.find_all('location'):
    province = location.find('province')
    if province and province.text == '강원도영서':
        children = location.find_all(recursive=False)
        for child in children:
            tag = child.name
            print(tag)
            text = re.sub('<[^>]+>', '', str(child))
            if tag == 'tmef': 
                text = '일시: ' + text
            elif tag == 'wf': 
                text = '날씨: ' + text
            elif tag == 'tmn': 
                text = '최저기온: ' + text
            elif tag == 'tmx': 
                text = '최고기온: ' + text
            elif tag == 'rnst': 
                text = '강수확률: ' + text
            elif tag == "mode":
                text = ""
            print(text)
            print("-" * 50)
```
