```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
def ctage(age):
    cat = ""
    if age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'
    return cat

def ctmonth(month):
    cat = ""
    if month < 3 or month == 12: cat = "겨울"
    elif month < 6: cat = "봄"
    elif month < 9: cat = "여름"
    else: cat = "가을"
    return cat

#========================================================================================================================
np.random.seed(5)
d_gen = {0 : {"Kpop" : 5,"Ballade" : 5,"Dance": 6,"Trot": 2,"Hip/Rock": 1},
        1 : {"Kpop": 5,"Ballade": 5,"Dance": 6,"Trot": 2,"Hip/Rock" : 10}}
d_age = {'Child' : {"Kpop" : 20,"Ballade": 8,"Dance": 12,"Trot" : 2,"Hip/Rock": 6},
        'Teenager' : {"Kpop" : 22,"Ballade": 14,"Dance": 24,"Trot" : 4,"Hip/Rock": 24},
        'Student' : {"Kpop" : 24,"Ballade": 20,"Dance": 30,"Trot" : 6,"Hip/Rock": 20}, 
        'Young Adult' : {"Kpop" : 26,"Ballade": 26,"Dance": 42,"Trot" : 8,"Hip/Rock": 18},
        'Adult' : {"Kpop" : 22,"Ballade": 20,"Dance": 42,"Trot" : 12,"Hip/Rock": 8},
       'Elderly' : {"Kpop" : 14,"Ballade": 20,"Dance": 30,"Trot" : 17,"Hip/Rock": 0}}
d_chan_22 = {"MelOn" : {"Kpop" : 11.91,"Ballade": 16.92,"Dance": 1,"Trot" : 0.3135,"Hip/Rock": 10.03},
            "Ginie Music" : {"Kpop" : 13.62,"Ballade": 16.87,"Dance": 2,"Trot" : 0.3245,"Hip/Rock": 10.38},
            "BUGS" : {"Kpop" : 11.50,"Ballade": 17.89,"Dance": 1,"Trot" : 0.3834,"Hip/Rock": 10.86},
            "Naver Music" :  {"Kpop" : 18.70,"Ballade": 17.36,"Dance": 1,"Trot" : 0.6012,"Hip/Rock": 11.35},
            "Flo" : {"Kpop" : 17.60,"Ballade": 22.88,"Dance": 1,"Trot" : 0.0044,"Hip/Rock": 13.20}}
d_chan_23 = {"MelOn" : {"Kpop" : 11.91,"Ballade": 16.92,"Dance": 1,"Trot" : 20,"Hip/Rock": 10.03},
            "Ginie Music" : {"Kpop" : 13.62,"Ballade": 16.87,"Dance": 2,"Trot" : 20,"Hip/Rock": 10.38},
            "BUGS" : {"Kpop" : 11.50,"Ballade": 17.89,"Dance": 1,"Trot" : 20,"Hip/Rock": 10.86},
            "Naver Music" :  {"Kpop" : 18.70,"Ballade": 17.36,"Dance": 1,"Trot" : 20,"Hip/Rock": 11.35},
            "Flo" : {"Kpop" : 17.60,"Ballade": 22.88,"Dance": 1,"Trot" : 20,"Hip/Rock": 13.20}}
d_season = {"봄" : {"Kpop" : 15,"Ballade": 16,"Dance": 7,"Trot" : 9,"Hip/Rock": 9},
           "여름" : {"Kpop" : 12,"Ballade": -2,"Dance": 30,"Trot" : 15,"Hip/Rock": 15},
           "가을" : {"Kpop" : 16,"Ballade": 26,"Dance": 21,"Trot" : 6,"Hip/Rock": 6},
           "겨울" : {"Kpop" : 12,"Ballade": 20,"Dance": -15,"Trot" : 12,"Hip/Rock": 12}}


#=========================================================================================================================

customer_id = np.arange(10000)
gender = np.random.randint(0,2,10000)
age = np.random.randint(10,81,10000)
age_cat = []
i = 0
while i < 10000:
    age_cat.append(ctage(age[i]))
    i += 1
region = np.random.randint(1,6,10000)

pdt = ["Kpop","Ballade","Dance","Trot","Hip/Rock"]
pdt_p22 = [0.3, 0.3, 0.3, 0.04, 0.06]
pdt_p23 = [0.2, 0.2, 0.2, 0.28, 0.12]
pdt_22 = [np.random.choice(pdt, p=pdt_p22) for _ in range(10000)]
pdt_23 = [np.random.choice(pdt, p=pdt_p23) for _ in range(10000)]
price_22 = np.random.randint(1000,50001,5)
pa_dic_22 = {"Kpop" : price_22[0], "Ballade" : price_22[1], "Dance" : price_22[2], "Trot" : price_22[3], "Hip/Rock" : price_22[4]}
pa_dic_23 = {"Kpop" : price_22[0]+np.random.randint(-1000,1000)
             , "Ballade" : price_22[1]+np.random.randint(-1000,1000)
             , "Dance" : price_22[2]+np.random.randint(-1000,1000)
             , "Trot" : price_22[3]+np.random.randint(-1000,1000)
             , "Hip/Rock" : price_22[4]+np.random.randint(-1000,1000)}
i = 0
pa_22 = []
while i < 10000:
    pa_22.append(pa_dic_22.get(pdt_22[i]))
    i += 1
i = 0
pa_23 = []
while i < 10000:
    pa_23.append(pa_dic_23.get(pdt_23[i]))
    i += 1
  

char = np.random.randint(1,6,10000)
chad = {1 : "MelOn",
          2: "Ginie Music",
          3: "BUGS",
          4: "Naver Music",
          5: "Flo"}
channel = []
i = 0
while i < 10000:
    channel.append(chad.get(char[i]))
    i += 1

t_22 = np.random.randint(1,13,10000)
t_23 = np.random.randint(1,13,10000)
season_22 = []
season_23 = []
i = 0
while i < 10000:
    season_22.append(ctmonth(t_22[i]))
    i += 1
i = 0
while i < 10000:
    season_23.append(ctmonth(t_23[i]))
    i += 1

                   
tr_y = np.random.randint(1,11,10000)
    
q_22 = []
q_23 = []
i = 0
while i < 10000:
    a = np.round(np.clip(np.random.normal(d_gen[gender[i]][pdt_22[i]] + d_age[age_cat[i]][pdt_22[i]] + 
                                          d_chan_22[channel[i]][pdt_22[i]] + d_season[season_22[i]][pdt_22[i]]
                                          - (pa_22[i]/3000)
                                 ,1,1),0,np.inf))
    q_22.append(int(a))
    i += 1
i = 0
while i < 10000:
    a = np.round(np.clip(np.random.normal(d_gen[gender[i]][pdt_23[i]] + d_age[age_cat[i]][pdt_23[i]] + 
                                          d_chan_23[channel[i]][pdt_23[i]] + d_season[season_23[i]][pdt_23[i]]
                                          - (pa_23[i]/3000)
                                 ,1,1),0,np.inf))
    q_23.append(int(a))
    i += 1








columns = ['id','gender','age','region','pdt_22',"pdt_23",'price_22',"price_23",'qty_22',"qty_23",'time_22',"time_23",
          "channel","tr_y","pur_22","pur_23","sales"]
data = {'id' : customer_id,'gender': gender,'age' : age,"age_cat" : age_cat,'region' : region,'pdt_22' : pdt_22,"pdt_23" : pdt_23,
        'price_22' : pa_22,"price_23" : pa_23,'qty_22' : q_22,"qty_23" : q_23, 
        'month_22' : t_22,"month_23" : t_23, "season_22" : season_22, "season_23" : season_23, "channel" : channel,"tr_y" : tr_y}





```

```python
df = pd.DataFrame(data)
df = df.set_index("id")
df["pur_22"] = df["price_22"] * df["qty_22"]
df["pur_23"] = df["price_23"] * df["qty_23"]
df["gender"].replace([0,1],["female","male"], inplace = True)
df
```
