### 과제1_0426
아래 text에서 ["file2.pdf", "file3.docx", "file4.xlsx"]만 출력하시오

```python
text = "file1.txt file2.pdf file3.docx file4.xlsx"
pattern = r"\b\w+\.(?!txt\b)\w+\b"
re.findall(pattern,text)
```

### 과제2_0426
아래 email에서 example만 출력하시오

```python
email = "john.doe@example.com"
pattern = r"(?<=@)\w+"
re.search(pattern,email).group()
```

### 과제3_0426
창의적인 사용자 함수를 생성하여 sub 메서드의 매개변수로 적용하는 사례를 작성하시오.

```python
def samplef(match):
    match = "red"
    return match

text = "I'll choose green one."
p = re.compile("green")
p.sub(samplef, text)
```

### 과제4_0426
네이버 사이트에서 아래 query_list를 검색할 수 있는 url을 생성하시오. (퍼센트 인코딩 변환)

query_list = ["파이썬", "웹 크롤링", "빅데이터", "python"]'

https://search.naver.com./search.naver?query=

```python
query_list = ["파이썬", "웹 크롤링", "빅데이터", "python"]
URL = "https://search.naver.com./search.naver?query="
url_list = []

for i in query_list:
    a = urllib.parse.quote(i)
    url_list.append(URL+a)
    
url_list
```

### 과제5_0426
https://www.daum.net/ 에서 아래 방법을 이용해서 한글 데이터만 출력하시오.

- 표준모듈을 이용
- 외부 라이브러리를 이용

```python
URL = "https://www.daum.net/"
response = urllib.request.urlopen(URL).read()

response = response.decode("utf-8")

re.findall("[가-힣]+", response)
```

```python
response = requests.get(URL)
html = response.content

soup = BeautifulSoup(html, "html.parser")
text = soup.get_text()
result = re.findall("[\u3131-\uD79D]+", text)

result
```
