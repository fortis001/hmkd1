```python
import random

foodIndex = ["햄버거","탕수육","피자","치킨","자장면","짬뽕","족발","라면","스테이크","파스타","샌드위치","타코","카레",
             "샐러드","떡볶이","초밥","칼국수","갈비찜","김치찌개","삼겹살","돈가스","볶음밥","불고기","브리또",
             "김밥","파전","해물찜","갈비탕","감자탕","떡갈비","국밥","찜닭"]

nextList = []
run = True
rounds = {"32" : 32, "16" : 16, "8" : 8, "1" : 32, "2" : 16, "3" : 8}



print("="*100)
print("<<<점심메뉴 이상형 월드컵>>>\n")
print("총 32개의 메뉴가 있습니다. 중간에 0을 입력하면 종료합니다.")
print("1. 32강\n2. 16강\n3. 8강")
print("="*100)


while True:
    select = input()
    if select in rounds:
        foodIndex = random.sample(foodIndex, rounds.get(select))
        print(f"{rounds.get(select)}강으로 시작합니다.")
        break
    elif select == "0":
        run = False
        break
    else:
        print("잘못된 입력입니다. 다시 입력해주세요.")

while len(foodIndex)+len(nextList) != 1 and run:
    Round = len(foodIndex)
    print(f'{Round}강')
    while len(foodIndex) != 0 and run:
        random.shuffle(foodIndex)
        a = foodIndex.pop()
        b = foodIndex.pop()
        print("-"*100)
        print(a, "vs", b)
        while 1:
            choose = input("1번 2번 중에 선택하세요!!\n>")
            if choose == "1" or choose == a:
                nextList.append(a)
                break
            elif choose == "2" or choose == b:
                nextList.append(b)
                break
            elif choose == "0":
                run = False
                break
            else:
                print('다시 입력해 주세요')
    foodIndex = nextList
    nextList = []



if run == False:
    print("종료되었습니다.")
else:
    print(f"우승은 {foodIndex}입니다!!!")
```
