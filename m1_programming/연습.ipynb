```python
import random

def upgrade_item(item_level):
    upgrade_chance = 0.7 - (item_level / 100)  # 아이템 강화 확률 계산
    choice = input("강화를 하시겠습니까? (y/n) ")
    if choice == "y":
        if random.random() < upgrade_chance:  # 랜덤으로 강화 성공 여부 결정
            item_level += 1
            print("아이템 강화 성공! 레벨 업! 현재 레벨", item_level)
        else:
            if item_level >= 6:
                item_level -= 1
                print("강화 실패 5강 이상입니다. 강화 수치가 1 감소합니다.")
                
            else:
                print("아이템 강화 실패...")
    elif choice == "n":
        print("강화를 취소했습니다.")
    else:
        print("잘못된 입력입니다. 다시 시도하세요.")
    
    return item_level

item_level = 1  # 초기 아이템 레벨
for i in range(10):  # 강화 시도 횟수
    item_level = upgrade_item(item_level)

print("최종 강화 결과", item_level,"강")
```

```python
import random

def upgrade_item():
    item_level = 1  # 초기 아이템 레벨
    while True:
        if 3 < item_level < 10:
            uchan = round(0.9 - 0.1*(item_level-3),2)
        elif item_level == 10:
            uchan = 0.25
        elif item_level == 11:
            uchan = 0.15
        elif 11 < item_level:
            uchan = round(0.15 - 0.01*(item_level-11),2)
        else:
            uchan = 1  # 아이템 강화 확률 계산
        choice = input(f"강화를 하시겠습니까? (y/n) 현재 레벨:{item_level}\n성공 확률:{uchan*100}%\n>")
        if choice == "y":
            if round(random.random(),2) < uchan:  # 랜덤으로 강화 성공 여부 결정
                item_level += 1
                print("아이템 강화 성공! 레벨 업! 현재 레벨", item_level)
            elif 11 > item_level > 5:
                item_level -= 1
                print("강화 실패 5강 이상입니다. 강화 수치가 1 감소합니다.")
            elif item_level > 10:
                D = random.randint(0,1)
                if D == 0:
                    print("아이템이 파괴되었습니다...")
                    item_level = 0
                    break
                else:
                    item_level -= 2
                    print("강화 실패, 강화 수치가 2 감소합니다.")        
            else:
                print("아이템 강화 실패...")
        elif choice == "n":
            print("강화를 취소했습니다.")
            break
        elif choice == "cheat":
            item_level = 14
        else:
            print("잘못된 입력입니다. 다시 시도하세요.")
    
    return item_level

    
item_level = upgrade_item()
print("최종 강화 결과", item_level,"강")
```

```python
import random

class ItemUpgrader:
    def __init__(self):
        self.item_level = 1
        
    def get_upgrade_chance(self):
        if 3 < self.item_level < 10:
            return round(0.9 - 0.1*(self.item_level-3),2)
        elif self.item_level == 10:
            return 0.25
        elif self.item_level == 11:
            return 0.15
        elif 11 < self.item_level:
            return round(0.15 - 0.01*(self.item_level-11),2)
        else:
            return 1
    
    def upgrade_item(self):
        while True:
            uchan = self.get_upgrade_chance()
            choice = input(f"강화를 하시겠습니까? (y/n) 현재 레벨:{self.item_level}\n성공 확률:{uchan*100}%\n>")
            if choice == "y":
                if round(random.random(),2) < uchan:
                    self.item_level += 1
                    print("아이템 강화 성공! 레벨 업!")
                elif 11 > self.item_level > 5:
                    self.item_level -= 1
                    print("강화 실패 5강 이상입니다. 강화 수치가 1 감소합니다.")
                elif self.item_level > 10:
                    D = random.randint(0,1)
                    if D == 0:
                        print("아이템이 파괴되었습니다...")
                        self.item_level = 0
                        break
                    else:
                        self.item_level -= 2
                        print("강화 실패, 강화 수치가 2 감소합니다.")        
                else:
                    print("아이템 강화 실패...")
            elif choice == "n":
                print("강화를 취소했습니다.")
                break
            elif choice == "cheat":
                self.item_level = 14
            else:
                print("잘못된 입력입니다. 다시 시도하세요.")
        
        print("최종 강화 결과", self.item_level,"강")                 
        return self.item_level
    



    
```
