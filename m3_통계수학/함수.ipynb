#### 역함수
- 어떤 함수의 입력/출력 관계와 정반대의 입출력 관계를 갖는 함수
- 원래의 함수에서 x축과 y축이 바뀐 것이므로 y = x가 나타내는 직선을 대칭축으로 대칭인 함수의 그래프

```python
from matplotlib import rc

rc('font',family='Malgun Gothic')  # 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 부호 출력 설정

def f1(x):
    return x**2
def f1inv(x):
    return np.sqrt(x)

x = np.linspace(0,3,300)
plt.figure(figsize = (10,6))
plt.plot(x, f1(x), "r-", label = "함수 $f(x) = x^2$")
plt.plot(x, f1inv(x), "b-", label = "역함수 $f^{-1}(x) = \sqrt{x}$")
plt.plot(x,x, "g--")
plt.axis("equal")
plt.xlim(0,2)
plt.ylim(0,2)
plt.legend()
plt.title("역함수의 그래프")
```

#### 지수함수
- 정해진 어떤 값을 입력값으로 거듭제곱하여 출력을 구하는 함수
- 거듭제곱할 때 아래에 있는 수를 밑이라 하고, 밑을 오일러 수 e(약 2.718)로 하여 거듭제곱 하는 함수를 지수함수라고 함.
- $y = e^x$, y = exp(x) = exp x로 표시
- 로그함수의 밑으로 자연로그함수 ln을 사용하는데, 이때의 밑이 오일러 상수 e이다.
