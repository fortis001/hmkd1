#### 과제 2_0421
위 그래프를 x를 400개로 해서 다시 표현하시오.

```python
def f(x):
    return x**3 - 3*x**2 + x
```

```python
x = np.linspace(-1, 3, 400)
y = f(x)
plt.plot(x, y, color = "limegreen")
```
