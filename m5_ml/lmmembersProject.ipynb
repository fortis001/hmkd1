```python
def getQ(date):
    cat = ''
    if date <= -1: cat = 'Unknown'
    elif date < 20140401: cat = '2014 1분기'
    elif date < 20140701: cat = '2014 2분기'
    elif date < 20141001: cat = '2014 3분기'
    elif date < 20150101: cat = '2014 4분기'
    elif date < 20150401: cat = '2015 1분기'
    elif date < 20150701: cat = '2015 2분기'
    elif date < 20151001: cat = '2015 3분기'
    else : cat = '2015 4분기'
    
    return cat
```

```python
def getH(date):
    cat = ''
    if date <= -1: cat = 'Unknown'
    elif date < 20140701: cat = '2014 상반기'
    elif date < 20150101: cat = '2014 하반기'
    elif date < 20150701: cat = '2015 상반기'
    else : cat = '2015 하반기'
    
    return cat
```

```python
def total_pie(df,year):
    corp = ["A","B","C","D"]
    list1 = []
    for i in corp:
        list1.append(df[df["제휴사"] == i]["TOTAL"].sum())
    plt.figure(figsize=(10, 6))
    plt.pie(list1, labels=corp, autopct='%1.1f%%', startangle=90)

    # 제목 추가
    plt.title(f'L사 {year}년 매출 비율')
    plt.legend(loc = "best")
```

```python
def df_to_lineplot(df,corp):
    df_A = df[df["제휴사"] == corp]
    df_A = df_A[["대분류코드","Q1_14","Q2_14","Q3_14","Q4_14"]].set_index("대분류코드")
    df_A = df_A.transpose()
    plt.title(f'{corp}사 14년 매출 추이')
    sns.lineplot(df_A)
    return df_A
```

```python
def get_outlier (df, column, weight = 1.5):
    fraud = df[column]
    quantile_25 = np.percentile(fraud.values, 25)
    quantile_75 = np.percentile(fraud.values, 75)
    
    iqr = quantile_75 - quantile_25
    iqr_weight = iqr * weight
    lowest_val = quantile_25 - iqr_weight
    highest_val = quantile_75 + iqr_weight
    
    outlier_index = fraud[(fraud < lowest_val) | (fraud > highest_val)].index
    return outlier_index
```
